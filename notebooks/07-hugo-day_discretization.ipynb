```python
import pandas as pd
```

```python
train = pd.read_csv('/home/hugo/data/DMC2017/train.csv', sep='|')
test = pd.read_csv('/home/hugo/data/DMC2017/class.csv', sep='|')
```

```python
train_day = pd.DataFrame(train[['lineID', 'day']])
test_day = pd.DataFrame(test[['lineID', 'day']])
```

```python
train_day['day_7'] = (train_day['day'] % 7)
train_day['day_10'] = (train_day['day'] % 10)
train_day['day_14'] = (train_day['day'] % 14)
train_day['day_28'] = (train_day['day'] % 28)
train_day['day_30'] = (train_day['day'] % 30)

test_day['day_7'] = (test_day['day'] % 7)
test_day['day_10'] = (test_day['day'] % 10)
test_day['day_14'] = (test_day['day'] % 14)
test_day['day_28'] = (test_day['day'] % 28)
test_day['day_30'] = (test_day['day'] % 30)
```

```python
test_day = test_day.drop('day', axis=1)
train_day = train_day.drop('day', axis=1)
```
